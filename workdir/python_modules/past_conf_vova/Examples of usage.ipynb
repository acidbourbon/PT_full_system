# Handling PASTTREC configurations

## Import all required libs

```python
import sys  
sys.path.insert(0, '../')

import pasttrec_ctrl as pt
import pasttrec_conf as pc

some_oep = "0x8012"
some_board = "0x1820"
```

## Creation of a configuration object

Next function will create a configuration object for specified tree of OEPs/FPGAs. This list can be generated using default functions: 
* pc.Helper.getConnectedMDCtree() - all connected MDC boards
* pc.Helper.getActiveMDCsubtree() - all connected and active MDC boards


```python
conf = pc.Configuration(mdc_tree=pc.Helper.getConnectedMDCtree())
print(conf)
```

or by hand:

```python
print(pc.Configuration(mdc_tree={'0x8012': ['0x1820', '0x1821']}))
```

## Navigation

### Through OEPs

To retrieve MDCTDC_OEP object for OEP with the given address, it's enough to just use this address as a key:

```python
print(conf[some_oep])
```

### Through FPGAs and PASTTREC

The same key access is available for every FPGA both in long:

```python
print(conf[some_oep][some_board])
```

and short form:

```python
print(conf[some_board])
```

And the same works for PASTTRECs

```python
print(conf[some_board][1])
```

## Changing settings

### Reset to default
All pasttrecs connected to OEP:

```python
conf[some_oep].resetConfig()
```

All pasttrecs connected to FPGA:

```python
conf[some_board].resetConfig()
```

Or specified PASTTREC:

```python
conf[some_board][1].resetConfig()
```

### Using setters

For changing any setting the better way is to use corresponding setter, as it will automaticly verify the data:

```python
conf[some_board].resetConfig()
print("Before changes was made:", conf[some_board], sep="\n")
conf[some_board][1].setGain(1) 
conf[some_board][1].setPt(15) 
conf[some_board][0].setThr(0)
conf[some_board][3].setThr(0) 
print("After changes was made:", conf[some_board], sep="\n")
```

### Using default configuration sets 

```python
conf[some_board][0].applyDefaultConfSet(pc.ConfStorage.memory_content_map["blue_settings_pt10_g1_thr127"])
print(conf[some_board])
```

### Baseline finder for specific TDC

```python
conf[some_board].found_baseline(scanning_time = 0.2)
print(conf[some_board])
```

### Automatic baseline finder for all TDC

```python
pt.reset_boards(["0x1820","0x1821"])
conf2 = pc.Configuration(mdc_tree = {"0x8012": ["0x1820", "0x1821"] })
conf2.found_all_baselines(plot=True)
print(conf2)
```

## Configuration test without uploading it to the onboard memory.
If allowed_tdc is empty - apply configuration to all boards.

```python
pt.reset_board("0x1820")
conf["0x1820"][1].setThr(0)
for past in conf["0x1821"]:
    past.setThr(0)
print(conf["0x1820"], conf["0x1821"], sep="\n")
conf.applyChanges(allowed_tdc=["0x1820", "0x1821"])
```

## Reading data back from real PASTTRECs.
For one PASTTREC:

```python
print(pc.MDCTDC_PT.read_configuration(some_board, pasttrec=0))
```

For whole TDC:

```python
print(pc.MDCTDC_FPGA.read_configuration(some_board))
```

## Upload configurations to FPGA flash

First of all, it's required to create an image structure of configuration on the file system. 

### Create an image in folder mytestimage

```python
conf.createImage( path="mytestimage")
print(conf[some_board])
```

This command creates the next file structure:
* mytestimage
    * 0x8001    - *oep*
        * 0x1801.hex  - *fpga*
        * 0x1800.hex
    * 0x8002
    * 0x8005
    ...
    
And inside each hex file there is a simple list of SC commands:
> 0xA0FF 0xFFFFFFFF <br>
> 0xA010 0x00050010<br>
> 0xA011 0x0005011E<br>
> 0xA012 0x0005022A<br>
> 0xA013 0x00050308<br>
> 0xA014 0x00050D05<br>
> ...

### Upload image to FPGA  

```python
pc.Configuration.flashImage(path="mytestimage", flash_settings="./flash_settings.pl -y", TDCs=[some_board])
pt.reset_boards([some_board])
```

# PASTTREC handling

## Slow control reset
To reset only PASTTREC SPI module without interruption anything else (will load settings from flash):

```python
pt.reset_board(some_board)
```

## Set value of specific PASTTREC register
Pasttrec IDs: [0 1 2 3]

```python
pt.spi_set_pasttrec_register(TDC_str=some_board,pasttrec=1,reg=0x3, data=0xFF)
```

## A baseline for the specific channel
Baseline range: -15:15

```python
pt.set_baseline(some_board,4,11)
```

## A baseline for the range of channels

```python
pt.set_all_baselines(TDC=some_board, channels=[1,2,3,5],values=[-15,0,15,13])
```

## Find all baselines for a board with noise method

```python
pt.found_baselines_for_board(board=some_board,scanning_time=1, plot=True, apply_res=True)
```

### Read PASTTREC registers
Raw data

```python
pt.reset_board("0x1820")
pt.read_PASTTREC_regs(TDC_str="0x1820",pasttrec=0)
```

### Load a default configuration set to the PASTTRECs from module memory

```python
pt.load_set_to_pasttrec(some_board, 0, addr=pc.ConfStorage.memory_content_map["black_settings_pt20_g1_thr127"]["addr"], len=pc.ConfStorage.memory_content_map["blue_settings_pt10_g1_thr127"]["len"])
print(pc.MDCTDC_FPGA.read_configuration(some_board))
```
