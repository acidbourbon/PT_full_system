```python
import hameg_trb 
import time

from matplotlib import pyplot as plt
%matplotlib inline
# plt.figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
# logging time in hours
time_range = 24
current_list = []
volt_list = []
current_list2 = []
volt_list2 = []
for o in range(0,8*60):
#         print(hameg_trb.get_curr(3), hameg_trb.get_curr(4))    
        time.sleep(60)
        current_list += [ hameg_trb.get_curr(3) ]
        volt_list += [ hameg_trb.get_volt(3) ]
        current_list2 += [ hameg_trb.get_curr(4) ]
        volt_list2 += [ hameg_trb.get_volt(4) ] 

        if o % 10 == 0:
        
            plt.scatter(range(0,len(current_list)),current_list)    
            plt.xlabel("minute")
            plt.ylabel("supply current [A] of all PASTTREC boards (5V) ")
            plt.show()

            plt.scatter(range(0,len(current_list2)),current_list2)        
            plt.xlabel("minute")
            plt.ylabel("supply current [A] of PASTTREC board 108 (5V) ")
            plt.show()

            plt.scatter(range(0,len(volt_list)),volt_list)
            plt.scatter(range(0,len(volt_list2)),volt_list2)        
            plt.xlabel("minute")
            plt.ylabel("supply voltage [V] of all / 108 PASTTREC board  ")
            plt.show()
```
